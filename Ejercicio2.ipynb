```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
def rgb_to_grayscale(image_rgb):
    return np.dot(image_rgb[..., :3], [0.299, 0.587, 0.114]).astype(np.uint8)
folder_name = 'colors'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
colors = {
    'red': (0, 0, 255),
    'green': (0, 255, 0),
    'blue': (255, 0, 0)
}
for color_name, bgr_value in colors.items():
    image = np.zeros((100, 100, 3), dtype=np.uint8)
    image[:] = bgr_value
    cv2.imwrite(f'{folder_name}/{color_name}.png', image)
for color_name in colors:
    image_path = f'{folder_name}/{color_name}.png'
    image_bgr = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
    grayscale_image = rgb_to_grayscale(image_rgb)
    print(f'Imagen: {color_name}.png - Valor del color (RGB): {image_rgb[0, 0]}')
    plt.subplot(1, 2, 1)
    plt.imshow(image_rgb)
    plt.title(f'{color_name} - Color')
    plt.axis('off')
    plt.subplot(1, 2, 2)
    plt.imshow(grayscale_image, cmap='gray')
    plt.title(f'{color_name} - Grayscale')
    plt.axis('off')
    plt.show()
```
